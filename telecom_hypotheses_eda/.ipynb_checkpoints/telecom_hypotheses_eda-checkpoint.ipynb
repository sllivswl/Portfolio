сообщение: 1 рубль

### Шаг 1. Импорт данных 

```python
from scipy import stats as st
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
calls = pd.read_csv('calls.csv')
internet = pd.read_csv('internet.csv')
messages = pd.read_csv('messages.csv')
tariffs = pd.read_csv('tariffs.csv')
users = pd.read_csv('users.csv')
```

```python
# Напишем функцию для ознакомления с данными:
def standart_opening(data):
    data.info()
    print('--------------------------------------')
    print('Дубликаты:')
    print (data.duplicated().sum(), 'дубликатов')
    print('--------------------------------------')
    print('Пропуски:')
    print (data.isna().mean())
    print('--------------------------------------')
    print('Начало таблицы')
    print (data.head())
    print('--------------------------------------')
    print('Конец таблицы')
    print (data.tail())

standart_opening(calls)
standart_opening(internet)
standart_opening(messages)
standart_opening(tariffs)
standart_opening(users)
```

Набор данных для анализа состоит из 5 таблиц разной величины. Пропусков и дубликатов нет.

### Шаг 2. Предобработка и подготовка данных

Для выполнения задания необходимо объединить данные. Чтобы рассчитать количество звонков, смс и интернета, потраченных каждым пользователем в течение месяца, я сделаю стобец `'month'` и сгруппирую все данные по нему и `user_id`.

Приведем данные к временному формату и выделим месяц регистрации пользователя:

```python
users.sample()
```

```python
users['reg_date'] = pd.to_datetime(users['reg_date'], format='%Y/%m/%d')
users['reg_month'] = pd.DatetimeIndex(users['reg_date']).month
users.sample()
```

Приведем данные в таблице `'calls'`к временному формату и выделим месяц совершения звонков пользователем:

```python
calls.sample()
```

```python
calls['call_date'] = pd.to_datetime(calls['call_date'], format='%Y/%m/%d')
calls['month'] = pd.DatetimeIndex(calls['call_date']).month
calls.rename(columns={'id': 'calls_id'}, inplace=True)
calls_data = calls.groupby(['user_id', 'month']).agg({'calls_id':'count', 'duration':'sum'}).reset_index()
calls_data.head()
```

Объединим получившиеся таблицы:

```python
data_1 = users.merge(calls_data, on='user_id', how='outer')
data_1.head()
```

Тоже самое сделаем с таблицей `'internet'`:

```python
internet.sample()
```

```python
internet['session_date'] = pd.to_datetime(internet['session_date'], format='%Y/%m/%d')
internet['month'] = pd.DatetimeIndex(internet['session_date']).month
internet.rename(columns={'id': 'internet_id'}, inplace=True)
internet_data = internet.groupby(['user_id', 'month']).agg({'internet_id':'count', 'mb_used':'sum'}).reset_index()
internet_data.head()
```

Присоединим полученную таблицу к предыдущим:

```python
data_2 = data_1.merge(internet_data, on=['user_id', 'month'], how='outer')
data_2.head()
```

Аналогично обработаем таблицу `'messages'` и соединим ее с остальными:

```python
messages.sample()
```

```python
messages['message_date'] = pd.to_datetime(messages['message_date'], format='%Y/%m/%d')
messages['month'] = pd.DatetimeIndex(messages['message_date']).month
messages.rename(columns={'id': 'messages_id'}, inplace=True)
messages_data = messages.groupby(['user_id', 'month']).agg({'messages_id':'count'}).reset_index()
messages_data.head()
```

```python
df = data_2.merge(messages_data, on=['user_id', 'month'], how='outer')
df.head()
```

Получилась таблица, которую удобно обрабатывать. Посмотрим, есть ли в ней пропуски и проверим типы данных:

```python
df.info()
```

`'Churn_date'` - это время окончания тарифа, пропуск в этом столбце означает, что им еще пользуются, но тип данных здесь надо привести к временному. Остальные нужно обработать. 

```python
df.isna().sum()
```

Для нас ключевым признаком является `'tariff'`. Удалим все строки с пропусками в этом столбце.

```python
df = df[df['tariff'].notnull()]
df.isna().sum()
```

Изучим пропуски в столбце `'month'`:

```python
df.loc[df['month'].isnull()]
```

Похоже, что 8 клиентов зарегистрировались, но не пользовались услугами - мы их удалим. 

```python
df = df[df['month'].notnull()]
df.isna().sum()
```

Остались пропущенные значения для столбцов `'internet_id'`, `'mb_used'`, `'messages_id'`. Возможно, данные остуствуют у клиентов, которые не использовали интернет и сообщения. Заменим пропуски на ноль. 

```python
df['internet_id'] = df['internet_id'].fillna(0)
df['mb_used'] = df['mb_used'].fillna(0)
df['messages_id'] = df['messages_id'].fillna(0)
df.isna().sum()
```

Приведем данные в столбце `'churn_date'` к временному формату:

```python
df['churn_date'] = pd.to_datetime(df['churn_date'], format='%Y/%m/%d')
```

```python
df.dtypes
```

```python
df[['month', 'calls_id', 'internet_id', 'messages_id']] = df[['month', 'calls_id', 'internet_id', 'messages_id']].astype('int')
df.dtypes
```

Я привела даты к соотвествующему формату. Месяц, количество звонков, сообщений и сессий интернета измеряются в целых числах. 

По условиям оператора все звонки меньше минуты округляются до 1 мин. Нужно привести данные в соответствие для расчетов по тарифам. 

```python
calls.head()
```

```python
calls.loc[(calls['duration'] > 0.00) & (calls['duration'] < 1.00), 'duration'] = 1
calls.query('0.00 < duration < 1.00')['duration'].count()
```

Теперь надо присоединить данные о тарифах, чтобы посчитать выручку. 

```python
len(df)
```

```python
df.head()
```

```python
# переименуем название столбца для присоединения
tariffs.rename(columns={'tariff_name': 'tariff'}, inplace=True)
df_t = df.merge(tariffs, on='tariff', how='outer')
len(df_t)
```

```python
df_t.head()
```

Добавим столбец с выручкой:

```python
def revenue(data):
    extra_min = 0
    extra_messages = 0
    extra_mb = 0
    
    if data['duration'] > data['minutes_included']:
        extra_min = (data['duration'] - data['minutes_included'])

    if data['messages_id'] > data['messages_included']:
        extra_messages = (data['messages_id'] - data['messages_included'])
        
    if data['mb_used'] > data['mb_per_month_included']:
        extra_mb = (data['mb_used'] - data['mb_per_month_included'])
        
    revenue = data['rub_monthly_fee'] \
    + extra_min * data['rub_per_minute'] + extra_messages \
    * data['rub_per_message'] + extra_mb / 1024 * data['rub_per_gb']
    return revenue

df_t['revenue'] = df_t.apply(revenue, axis=1)    
df_t.head()  
```

### Шаг 4. Проверка гипотез


- cредняя выручка пользователей тарифов «Ультра» и «Смарт» различается;


Для того, чтобы проверить эту гипотезу необходимо сопоставить средние значения выручки двух генеральных совокупностей. Прежде чем производить расчет я хочу выяснить, насколько данные по разным тарифам схожи по своему объему. 

```python
df.query('tariff == "smart"')['tariff'].count()
```

```python
df.query('tariff == "ultra"')['tariff'].count()
```

Выборка с пользователями тарифа "Смарт" занчительно больше, поэтому я укажу  `equal_var=False`.
Нулевая гипотеза будет звучать так: **"Средняя выручка пользователей тарифов «Ультра» и «Смарт» не различаются"**. 

```python
smart_revenue = smart_data['revenue']
ultra_revenue = ultra_data['revenue'
alpha = 0.05

results = st.ttest_ind(smart_revenue, ultra_revenue, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза: "Средняя выручка пользователей тарифов «Ультра» и «Смарт» не различаются" - отвергнута, следовательно гипотеза "Средняя выручка пользователей тарифов «Ультра» и «Смарт» различается" получает потдтверждение. 

- средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов;

Для проверки этой гипотезы необходимо подготовить соответствующие выборки по городу проживания пользователя и оценить их объем.

```python
df_t.sample()
```

```python
df_t.query('city == "Москва"')['city'].count()
```

```python
df_t.query('city != "Москва"')['city'].count()
```

Выборка с пользователями из Москвы почти в 5 раз меньше. Нулевая гипотеза будет звучать так: **"Средняя выручка пользователей из Москвы и не из Москвы не различается"**. 

```python
moscow_revenue = df_t.query('city == "Москва"')['revenue']
rest_revenue = df_t.query('city != "Москва"')['revenue']
alpha = 0.05

results = st.ttest_ind(moscow_revenue, rest_revenue, equal_var = False)

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Нулевая гипотеза: "Средняя выручка пользователей из Москвы и не из Москвы не различается" -  не была отвергнута.
Полученное значение `p-value` говорит о том, что хотя средняя выручка от пользователей в Москве и регионах различается,  с вероятностью в 30% такое или большее различие можно получить случайно. 
Это не позволяет делать вывод о что средняя выручка между Москвой и регионами не равны. Следовательно альтернативная  гипотеза "Средняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов" не получает потдтверждение.

#### *Таким образом, можно сделать вывод, что для компании «Мегалайн» выгоднее увеличивать количество клиентов тарифа «Ультра». Несмотря на то, что пользователи «Смарт» часто выходят за рамки лимитов, денег они приносят меньше. Статистически значимой разницы в выручке от клиентов из Москвы и других регионов нет.*
